```python
from bs4 import BeautifulSoup
import requests
import time
import datetime

import smtplib
```

```python
# connect to webste

url = 'https://www.walmart.com/ip/3-pack-George-Men-s-Big-Men-s-Cotton-Crewneck-T-Shirt-with-Long-Sleeves-Size-XS-3XL/18323170440?classType=VARIANT&athbdg=L1600'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36","Accept-Language": "en-US,en;q=0.9"}


page = requests.get(url, headers=headers)

soup1 = BeautifulSoup(page.content, 'html.parser')

soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')

title = soup2.find(id = 'main-title')

price = soup2.find('span', itemprop = 'price')

print(title)
print(price)
```

```python
title = title.get_text().strip()
price = price.get_text().strip()[1:]

print(title)
print(price)
```

```python
import datetime
today = datetime.date.today()
print(today)
```

```python
import csv

header = ['Title', 'Price', 'Date']
data = [title,price,today]

with open('WalmartWebScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
    
```

```python
def check_price():
    url = 'https://www.walmart.com/ip/3-pack-George-Men-s-Big-Men-s-Cotton-Crewneck-T-Shirt-with-Long-Sleeves-Size-XS-3XL/18323170440?classType=VARIANT&athbdg=L1600'

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36","Accept-Language": "en-US,en;q=0.9"}
    
    
    page = requests.get(url, headers=headers)
    
    soup1 = BeautifulSoup(page.content, 'html.parser')
    
    soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')
    
    title = soup2.find(id = 'main-title')
    
    price = soup2.find('span', itemprop = 'price')
    
    title = title.get_text().strip()
    price = price.get_text().strip()[1:]

    import datetime
    today = datetime.date.today()
    
    import csv
    header = ['Title', 'Price', 'Date']
    data = [title,price,today]
    
    with open('WalmartWebScraperDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)



```
